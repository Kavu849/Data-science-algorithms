```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
N = data.shape[0]
d = data.shape[1]
ads_sel = []
N_1 = [0] * d
N_0 = [0] * d

for n in range(N):
    ad = 0
    max_theta = 0
    for i in range(d):
        theta = random.betavariate(N_1[i] + 1, N_0[i] + 1)
        if (theta > max_theta):
            max_theta = theta
            ad = i
    ads_sel.append(ad)
    reward = data.values[n, ad]
    if reward == 1:
        N_1[ad] = N_1[ad] + 1
    else:
        N_0[ad] = N_0[ad] + 1
```

```python
plt.hist(ads_sel)
```
