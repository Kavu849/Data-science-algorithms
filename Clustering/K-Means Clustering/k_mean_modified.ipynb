```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
x = data.iloc[:,3:5].values
```

```python
# elbow method to find the optimal amount of clusters
```

```python
from sklearn.cluster import KMeans
wcss = []
for n in range(1,11):
    clust = KMeans(n_clusters = n, init = 'k-means++', random_state = 42)
    clust.fit(x)
    wcss.append(clust.inertia_)
plt.plot(range(1,11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clust = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = clust.fit_predict(x)
```

```python
print(y_kmeans)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(clust.cluster_centers_[:, 0], clust.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
